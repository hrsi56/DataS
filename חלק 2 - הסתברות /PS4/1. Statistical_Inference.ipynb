### **שיעור: הסקה סטטיסטית (Statistical Inference)**

---

### **מבוא להסכה סטטיסטית**

**הסקה סטטיסטית** היא תהליך ניתוח נתונים לצורך הסקת מסקנות על תהליך יצירת הנתונים (Data Generating Process - DGP).

- **למה זה חשוב?**  
  ללא הסקה סטטיסטית אנחנו פשוט "חיים בתוך הנתונים" ללא הבנה של המנגנונים שמאחוריהם. עם הסקה סטטיסטית, אנחנו יכולים לגלות תובנות חדשות על התהליך שיצר את הנתונים.

### **סוגי הסקה סטטיסטית:**

1. **אמידה נקודתית (Point Estimation):**  
   מציאת ערך שמייצג את ההערכה הטובה ביותר לפרמטר מסוים.

2. **אמידת אינטרוול (Interval Estimation):**  
   קביעת טווח ערכים (כגון רווח סמך) שבו נמצא הפרמטר בהסתברות מסוימת.

3. **בדיקת השערות (Hypothesis Testing):**  
   קבלת או דחיית השערה סטטיסטית לגבי פרמטרים באוכלוסייה.

### **דגימות מקריות עצמאיות (i.i.d):**

דגימות הן **מקריות ועצמאיות** אם:
- **עצמאות:** כל דגימה אינה תלויה בדגימות אחרות.
- **התפלגות זהה:** כל הדגימות מגיעות מאותה התפלגות.

### **דוגמה 1: הטלת קוביה הוגנת**

1. **מהי התוחלת של הטלת קוביה הוגנת?**$$
   E(X) = \frac{1+2+3+4+5+6}{6} = 3.5
   $$

```python
import numpy as np
import matplotlib.pyplot as plt

# פונקציה לחישוב ממוצעים עבור מספר חזרות וגודל דגימה
def simulate_dice_rolls(num_rolls, num_simulations):
    means = []
    for _ in range(num_simulations):
        rolls = np.random.randint(1, 7, num_rolls)
        means.append(np.mean(rolls))
    return means

# הגדרות
num_simulations = 1000
sample_sizes = [10, 100, 1000]

# הרצת סימולציות והצגת גרפים
plt.figure(figsize=(15, 5))
for i, size in enumerate(sample_sizes, 1):
    means = simulate_dice_rolls(size, num_simulations)
    plt.subplot(1, 3, i)
    plt.hist(means, bins=30, density=True, alpha=0.7, edgecolor='black')
    plt.axvline(3.5, color='red', linestyle='--', label='True Mean (3.5)')
    plt.title(f'Distribution of Sample Means\nSample Size = {size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

### **דוגמה 2: דגימות מהתפלגות בינומיאלית (Binomial Distribution)**

1. **מהי התוחלת של $( X \sim \text{Binomial}(10, 0.2) )$ ?**$$
   E(X) = n \cdot p = 10 \cdot 0.2 = 2
   $$

```python
def simulate_binomial_draws(num_draws, num_simulations, n=10, p=0.2):
    means = []
    for _ in range(num_simulations):
        draws = np.random.binomial(n, p, num_draws)
        means.append(np.mean(draws))
    return means

# הרצת סימולציות והצגת גרפים
plt.figure(figsize=(15, 5))
for i, size in enumerate(sample_sizes, 1):
    means = simulate_binomial_draws(size, num_simulations)
    plt.subplot(1, 3, i)
    plt.hist(means, bins=30, density=True, alpha=0.7, edgecolor='black')
    plt.axvline(2, color='red', linestyle='--', label='True Mean (2)')
    plt.title(f'Distribution of Binomial Sample Means\nSample Size = {size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

### **דוגמה 3: דגימות מהתפלגות מעריכית (Exponential Distribution)**

1. **מהי התוחלת של $( X \sim \text{Exponential}(2) )$ ?**$$
   E(X) = \frac{1}{\lambda} = \frac{1}{2} = 0.5
   $$

```python
def simulate_exponential_draws(num_draws, num_simulations, rate=2):
    means = []
    for _ in range(num_simulations):
        draws = np.random.exponential(scale=1/rate, size=num_draws)
        means.append(np.mean(draws))
    return means

# הרצת סימולציות והצגת גרפים
plt.figure(figsize=(15, 5))
for i, size in enumerate(sample_sizes, 1):
    means = simulate_exponential_draws(size, num_simulations)
    plt.subplot(1, 3, i)
    plt.hist(means, bins=30, density=True, alpha=0.7, edgecolor='black')
    plt.axvline(0.5, color='red', linestyle='--', label='True Mean (0.5)')
    plt.title(f'Distribution of Exponential Sample Means\nSample Size = {size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
