# י 📌 $Bootstrapping$ ובגינג ($Bagging$) בעצים י  

## **📌 י מבוא י**  
י בעת שימוש בעצים להסקה סטטיסטית או למידת מכונה, אנו עשויים להיתקל בבעיה של **שונות גבוהה** ($High Variance$).  י  
י זאת אומרת, אם נשנה מעט את נתוני האימון, המבנה של העץ עשוי להשתנות באופן משמעותי.  י  
י כדי להתמודד עם בעיה זו, אנו משתמשים ב- **$Bootstrap$ $Aggregation$ ($Bagging$)**, טכניקה שמקטינה שונות ומגבירה דיוק. י  

📌 **י מטרה:**  י  
י לשפר את הדיוק של עצים על ידי שילוב של **הרבה עצים שונים**, כל אחד מהם מאומן על **מדגם שונה של הנתונים**. י  

---

## **📌 י מהו $Bootstrapping$? י**  
י $Bootstrap$ היא שיטה סטטיסטית להערכת **אי-ודאות במודלים** באמצעות **דגימה מחדש ($Resampling$)** עם החזרה. י  

🔹 **י שלבים:** י  
1. י נניח כי יש לנו **מערך נתונים מקורי** $( Z = \{z_1, z_2, ..., z_n\} )$. י  
2. י יוצרים **($B$)** דגימות חדשות בגודל זהה למערך הנתונים המקורי **(עם החזרה)**. י  
3. י מאמנים את המודל **($B$)** פעמים על כל דגימה בנפרד. י  
4. י מנתחים את ההתפלגות של המדד הסטטיסטי $( S(Z) )$ על פני הדגימות. י  

📌 **י הערכת שונות במדד סטטיסטי $( S(Z) )$:** י  
י בבוטסטראפ ($Bootstrap$) אנו מעוניינים לאמוד את השונות ($Variance$) של מדד סטטיסטי מסוים $( S(Z) )$ – למשל, מדד ביצועים של עץ סיווג (כמו דיוק, $Recall$, $Precision$, או מדד אחר), על סמך דגימות חוזרות מהמדגם המקורי. י  

י הנוסחה הבסיסית להערכת השונות בנוסח הבוטסטראפ היא: י  
$$
\widehat{VAR} [S(Z)] 
= \frac{1}{B - 1} \sum_{b=1}^{B} \Bigl(S(Z^*_b) - \overline{S}^* \Bigr)^2
$$

---

## **📌 י מהו $Bagging$ וכיצד הוא עובד? י**  
י **$Bagging$** (קיצור של *$Bootstrap$ $Aggregation$*) היא טכניקה ל**שילוב מודלים** ($Ensemble$ $Learning$) המבוססת על רעיון הבוטסטראפ ($Bootstrap$). י  

### **📌 י איך זה עובד? י**  
1. י **דגימת בוטסטראפ ($Bootstrap$)**  י  
   - י יוצרים $B$ דגימות חדשות (כל אחת בגודל $n$), על ידי דגימה עם החזרה מתוך המדגם המקורי. י  
2. י **אימון מודלים שונים על כל דגימה**  י  
   - י על כל אחת מ- $B$ דגימות הבוטסטראפ מאמנים מודל עצמאי חדש (למשל, עץ החלטה). י  
3. י **איחוד התחזיות ($Aggregation$)**  י  
   - י אם המודלים הם **רגרסיה** – לוקחים את **ממוצע** התחזיות. י  
   - י אם המודלים הם **סיווג** – עושים **הצבעה רוב ($Majority$ $Vote$)**. י  

---

# **📌 י $Bagging$ למה השיטה טובה? י**  
## **📌 י 1. הקטנת שונות ($Variance$) י**  
### **📌 י הרעיון הכללי י**  
י עצים ($Decision$ $Trees$) לבדם נוטים לשונות גבוהה יחסית. כאשר מאמנים מספר רב של עצים על דגימות בוטסטראפ שונות וממזגים את התוצאות שלהם (למשל, על ידי הצבעת רוב בסיווג), השונות של התוצאה הסופית יורדת. י  

## **📌 י 2. מניעת $Overfitting$ י**  
י עץ החלטה בודד נוטה ל-$Overfitting$. על ידי אימון עצים רבים ושילובם יחד, ההסתמכות על "הטיות" של מדגם יחיד פוחתת, והממוצע (או הצבעת הרוב) מתקבל כיותר "כללי" ופחות מוטה לפרטים מקריים בנתונים. י  

## **📌 י 3. שמירה על הטיה ($Bias$) נמוכה י**  
י עצי החלטה הם לרוב מודלים עם הטיה נמוכה: הם מסוגלים להתאים את עצמם היטב למבנה האמיתי של הנתונים. לכן, גם כאשר משכללים אותם בשיטת $Bagging$, הטיה זו לרוב **לא** גדלה בצורה משמעותית. י  

---

## **📌 י סיכום קצר י**  
י 1. **הקטנת שונות ($Variance$)** – שילוב של עצים שונים (שנוצרו מדגימות בוטסטראפ) בולם את השפעות הרעש הנמצא בכל מדגם בודד. י  
י 2. **מניעת $Overfitting$** – עצים בודדים נוטים להתאים יתר על המידה; ממוצע/הצבעת רוב של עצים רבים הופכת את המודל הסופי לעמיד יותר מפני אפיון יתר. י  
י 3. **שמירה על הטיה ($Bias$) נמוכה** – עצי החלטה הם מודלים בעלי הטיה נמוכה יחסית, ושיטת $Bagging$ לרוב לא מגדילה אותה באופן משמעותי. י  

י שילוב התכונות הללו הופך את באגינג למתודה חזקה מאוד באנסמבל ($Ensemble$ $Learning$), בייחוד כאשר מתמודדים עם מצבים של הרבה רעש ואי-ודאות בנתונים. י


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

# יצירת נתונים סינתטיים
np.random.seed(42)
X = np.random.rand(200, 1) * 10
Y = X[:, 0] * np.sin(X[:, 0]) + np.random.randn(200) * 2

# חלוקת הנתונים לסט אימון ובדיקה
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# אימון עץ רגרסיה בודד
tree_reg = DecisionTreeRegressor(max_depth=3)
tree_reg.fit(X_train, Y_train)
Y_pred_tree = tree_reg.predict(X_test)

# אימון מודל Bagging עם 50 עצים
bagging_reg = BaggingRegressor(estimator=DecisionTreeRegressor(max_depth=3), n_estimators=50, random_state=42)
bagging_reg.fit(X_train, Y_train)
Y_pred_bagging = bagging_reg.predict(X_test)

# הצגת התוצאות
plt.figure(figsize=(10, 6))
plt.scatter(X_test, Y_test, color="gray", alpha=0.5, label="Actual Data")
plt.scatter(X_test, Y_pred_tree, color="red", label="Single Tree Predictions")
plt.scatter(X_test, Y_pred_bagging, color="blue", label="Bagging Predictions")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Comparison: Single Decision Tree vs. Bagging")
plt.legend()
plt.show()
```

# י הסבר תיאורטי על הגרף של $Single$ $Decision$ $Tree$ מול $Bagging$ י  

## **📌 י מה מוצג בגרף? י**  
י בגרף מוצגים שלושה מרכיבים עיקריים: י  
1. **י נקודות אפורות $(Actual$ $Data)$** - נקודות הבדיקה $(Test$ $Set)$ והתוצאה בפועל $(Y$ האמיתיים$)$. י  
2. **י נקודות אדומות $(Single$ $Tree$ $Predictions)$** - התחזיות של עץ רגרסיה יחיד $(max_depth=3)$. י  
3. **י נקודות כחולות $(Bagging$ $Predictions)$** - התחזיות של אנסמבל $(Ensemble)$ של 50 עצי רגרסיה $(max_depth=3)$ בשיטת $Bagging$. י  

## **📌 י השוואת ביצועי עץ יחיד מול $Bagging$ י**  
- **י עץ רגרסיה יחיד (אדום)**: י  
  - נוטה לספק תחזיות "גסות" עקב עומק מוגבל. י  
  - עלול לטעות טעות יחסית גדולה באזורים מסוימים, במיוחד אם הנתונים שם רועשים. י  
- **י $Bagging$ (כחול)**: י  
  - משלב מספר רב של עצים, שכל אחד מאומן על דגימת $Bootstrap$ שונה. י  
  - התחזיות "מתייצבות" יותר ופחות רגישות לרעש מקומי או הטיות במדגם. י  
  - התוצאה: עקומה שקרובה יותר לפונקציה האמיתית, עם שונות $(Variance)$ מופחתת. י  

## **📌 י מדוע ניתן לצפות לשיפור בביצועים? י**  
1. **י הקטנת שונות $(Variance)$**  י  
   - ממוצע מספר רב של עצים מפחית שגיאות שנובעות מרעש נתונים. י  
2. **י מניעת $Overfitting$**  י  
   - שילוב עצים מאומנים על דגימות שונות הופך את המודל הסופי ליותר כללי ועמיד. י  
3. **י שמירה על $Bias$ נמוך**  י  
   - עצי החלטה נחשבים למודלים בעלי $Bias$ נמוך, ו-$Bagging$ לא מעלה אותו משמעותית. י  

---

# **📌 י השפעת $Bagging$ על עצי החלטה י**  

## **📌 י מה ניתן לראות? י**  
- 🔴 **י הנקודות האדומות** → חיזוי של עץ יחיד $(נוטה ל-Overfitting)$. י  
- 🔵 **י הנקודות הכחולות** → חיזוי עם **$Bagging$** $(יותר יציב ומדויק)$. י  

## **📌 י כיצד מודדים דיוק של $Bagging$? י**  

🔹 **י יתרון מרכזי של $Bagging$**: י  
ניתן לאמוד את **שגיאת הבדיקה $(Test$ $Error)$** **ללא קרוס-ולידציה**. י  

## **📌 י רעיון $Out-Of-Bag$ $(OOB)$ י**  
- לכל עץ יש **נתונים שלא נבחרו** במהלך **$Bootstrap$**. י  
- ניתן **להעריך ביצועים** של המודל על **תצפיות אלו**. י  

### **📌 י נוסחת חישוב $OOB$ $MSE$**: י  
$$
\hat{MSE}_{boot} = \frac{1}{n} \sum_{i=1}^{n} \frac{1}{|C^{-i}|} \sum_{b \in C^{-i}} (y_i - \hat{f}^{*b}(x_i))^2
$$

📌 **י משמעות:** י  
- במקום לבצע **$Cross-Validation$ יקרה**, ניתן להשתמש ב-**$OOB$ $Error$** להערכת ביצועים **אמינה וזולה יותר**. י  

---

## **📌 י יתרונות וחסרונות של $Bagging$ י**  
| **י יתרון י** | **י חיסרון י** |
|-----------|-----------|
| מפחית שונות **$(High$ $Variance)$** | אינו מצמצם **$Bias$ גבוה** |
| משפר דיוק על פני מודל יחיד | צורך **חישובי גבוה יותר** |
| עובד טוב עם **עצים לא מגוזמים** | פחות יעיל על **מודלים ליניאריים** |

📌 **י $Bagging$ עובד היטב כאשר העצים הבודדים הם בעלי שונות גבוהה אך בעלי $Bias$ נמוך! י**  

---

## **📌 י השוואה בין $Bagging$ לעץ יחיד י**  
| **י מאפיין י** | **י עץ בודד $(Single$ $Tree)$ י** | **י $Bagging$ $(Bootstrap$ $Aggregation)$ י** |
|------------|----------------------|--------------------------|
| **י שונות $(Variance)$** | גבוהה **$(רגישות לנתונים)$** | נמוכה **$(ממוצע עצים מפחית שונות)$** |
| **י הטיה $(Bias)$** | משתנה לפי גודל העץ | אינה משתנה משמעותית |
| **י דיוק י** | לעיתים לא יציב | לרוב **מדויק יותר** |
| **י $Overfitting$** | אפשרי בעץ עמוק | **מופחת** |
| **י חישוביות י** | נמוכה | **גבוהה יותר** |

📌 **י $Bagging$ אידיאלי כאשר יש צורך במודל גמיש אך יציב, במיוחד בעצי החלטה לא מגוזמים. י**  

---

## **📌 י סיכום י**  
✔ **י $Bootstrapping$ מאפשר יצירת דגימות חדשות על ידי **דגימה עם החזרה**. י  
✔ **י $Bagging$ משתמש ב-$Bootstrapping$ כדי לאמן **מספר עצים** ולמזג אותם **לשיפור הדיוק**. י  
✔ **י $Out-of-Bag$ $Error$ מספק דרך **להערכת ביצועים ללא קרוס-ולידציה**. י  
✔ **י $Bagging$ מפחית שונות ומשפר דיוק**, אך **דורש משאבים חישוביים גבוהים יותר**. י  

📌 **י באמצעות $Bagging$ ניתן לשפר משמעותית את הביצועים של עצים ולעבור למודלים מורכבים יותר כמו $Random$ $Forests$! 🚀 י**

1) Dimensionality Reduction

א. סיכום קצר של השיטה

מה זה?
“צמצום ממדים” מתייחס למשפחה של טכניקות שמטרתן לצמצם את מספר התכונות (Features) או הממדים בדאטה, תוך שמירה מרבית על המידע הרלוונטי. דוגמאות נפוצות:
	•	PCA (Principal Component Analysis): מציאת צירופים ליניאריים של מאפיינים שמסבירים את עיקר השונות בנתונים.
	•	t-SNE או UMAP: בעיקר לצורכי ויזואליזציה בממדים נמוכים.
	•	Autoencoders (ברשתות נוירונים): מפותחות גם הן לצמצום ממדים ולמידת ייצוגים דחוסים.

ב. מדוע המודל הזה טוב (יתרונותיו)
	1.	התמודדות עם Curse of Dimensionality: כאשר מספר התכונות גבוה מאוד ביחס למספר הדוגמאות, ביצועי מודלי ML עלולים להיפגע, ו-Dimensionality Reduction יכול לעזור להבליט את המידע המשמעותי ביותר.
	2.	פשטות חישובית: מודל שהריץ צמצום ממדים יכול להשתמש בפחות תכונות בפועל – חסכון בזמן ובזיכרון.
	3.	הפחתת רעש: לעיתים הצמצום “מעלים” תכונות שהן בעיקר רעש, ובכך משפר את הכללת המודל.

ג. יתרונות של XGBoost לעומת Dimensionality Reduction
	1.	מודל חיזוי מלא: Dimensionality Reduction אינו “מודל חיזוי” בפני עצמו, אלא פרה-פרוססינג (Preprocessing) או שלב עזר. XGBoost כן מהווה אלגוריתם למידה מפוקח (Supervised) שמוציא תחזיות ישירות.
	2.	מנגנון מובנה ל-Feature Importance: XGBoost יכול בעצמו להפיק מדדי חשיבות לכל תכונה, ואף לבצע “Column Subsampling” כדי להתמודד עם מספר תכונות גדול, במקום למזער ממדים בצורה מפורשת.
	3.	יכולות הקבלה (Scalability): XGBoost מותאם לדאטה גדול מאוד, ומספק אופטימיזציות לזמן ריצה. בעוד Dimensionality Reduction עשוי לדרוש חישובים מורכבים (למשל SVD מלא ב-PCA) שלא תמיד עומדים בקנה אחד עם סקיילים גדולים.

ד. דמיון ושוני מול XGBoost
	•	דמיון: שני הכלים יכולים להוות חלק מ-Pipeline ב-ML, כשלעיתים מבצעים צמצום ממדים לפני אימון מודל (כולל XGBoost).
	•	שוני:
	1.	מטרה: מטרת Dimensionality Reduction היא להקטין את מרחב התכונות, בעוד XGBoost הוא מודל חיזוי המבוסס על עצים.
	2.	סוג אלגוריתם: Dimensionality Reduction אינו בהכרח מפיק “Predictive Model”; XGBoost הוא מודל ניבוי לכל דבר (סיווג, רגרסיה וכו’).
	3.	רגולריזציה מול צמצום ממדים: בעוד צמצום ממדים הוא אחת הדרכים להתמודד עם עודף פרמטרים, XGBoost מתמודד עם מורכבות יתר באמצעות רגולריזציה על העצים, סבסמפול, ועוד.

⸻

2) Bootstrapping

א. סיכום קצר של השיטה

מה זה?
Bootstrapping היא טכניקת דגימה סטטיסטית. הרעיון הבסיסי:
	•	לקחת דגימות חוזרות (Sample with replacement) מהסט המקורי, באותו גודל כמו הסט עצמו, אך עם חזרות אקראיות.
	•	משמשת בין היתר לאמידת סטיית תקן של אומדנים, בניית רווחי סמך, או כבסיס לשיטות Ensemble כמו Bagging.

ב. מדוע השיטה הזו טובה (יתרונותיה)
	1.	אמידה סטטיסטית גמישה: מאפשרת להעריך פרמטרים כמו ממוצע או מדד סטטיסטי אחר כאשר ההנחות הקלאסיות (למשל התפלגות נורמלית) לא בהכרח מתקיימות.
	2.	בסיס לשיטות Ensemble: Bootstrapping הוא היסוד ליצירת סטים שונים של אימון ולבניית מודלים עצמאיים שבסוף מקבלים “Vote” או ממוצע.

ג. יתרונות של XGBoost על פני Bootstrapping
	1.	XGBoost כאלגוריתם שלם: Bootstrapping בפני עצמו הוא תהליך דגימה, לא מודל למידה. XGBoost הוא Pipeline שלם משלב אובדן, גרדיאנטים, בניית עצים ורגולריזציה.
	2.	Boosting במקום “רק דגימה”: בניגוד לשימוש ב-Sampling כדי ליצור מודלים עצמאיים ואז למזג אותם (כמו ב-Bagging), Boosting פועל רספונסיבית לשגיאות המודל הנוכחי. הוא לא רק משנה את מדגם האימון, אלא גם משנה את הפונקציה הנלמדת בכל שלב לפי השגיאה.

ד. דמיון ושוני מול XGBoost
	•	דמיון:
	•	שני המושגים יכולים להופיע בעולם ה-Ensemble. XGBoost גם עושה Subsampling (דגימה של חלק מהדאטה) בשלבים מסוימים, אם כי לא תמיד בשיטת “With Replacement”.
	•	שוני:
	1.	מטרה ושימוש: Bootstrapping הוא כלי סטטיסטי למגוון מטרות (אמידה, רווחי סמך), בעוד XGBoost הוא מודל חיזוי מסוג Gradient Boosted Trees.
	2.	תהליך: Bootstrapping לא מחייב עדכון הדרגתי של המודל לפי השגיאה, אלא לרוב יצירת מדגמים שונים במקביל. XGBoost כן בנוי על תיקון שגיאות של השלבים הקודמים (Boosting).

⸻

3) Bagging

א. סיכום קצר של השיטה

מה זה?
Bagging (Bootstrap Aggregating) הוא אלגוריתם Ensemble שבו מאמנים כמה מודלים (נניח עצים), כל אחד על מדגם Bootstrap עצמאי של הדאטה (Sample with replacement), ואז מאגדים (Aggregating) את תחזיותיהם על ידי ממוצע (ברגרסיה) או הצבעה (בסיווג).

ב. מדוע המודל הזה טוב (יתרונותיו)
	1.	הקטנת שונות (Variance): על ידי הרצת מודלים רבים על מדגמים שונים מהדאטה, הטעויות האקראיות של כל מודל מתקזזות בממוצע הסופי.
	2.	פשטות: שקל ליישם; אפשר לקחת כל אלגוריתם בסיסי (כמו עץ החלטה), ולבצע עליו Bagging.
	3.	Scalability: ניתן לאמן מודלים בנפרד (במקביל) ולשלב את התוצאות.

ג. יתרונות של XGBoost על פני Bagging
	1.	למידה אדפטיבית: בעוד Bagging בונה מודלים במקביל על דגימות Bootstrap, XGBoost בונה עץ אחרי עץ, כאשר כל עץ מתרכז בשגיאות של הקודם (Gradient Boosting). זה מוביל לעיתים קרובות למודל מדויק יותר מאשר “רק” ממוצע עצים עצמאיים.
	2.	רגולריזציה מתקדמת: XGBoost כולל פרמטרים כמו \gamma, \lambda, ו-\alpha כדי להעניש עצים מורכבים. Bagging (בעצמו) לא מבצע רגולריזציה על מורכבות המודל, מעבר לעובדה שכל מודל בסיסי בפני עצמו עשוי להיות מוגבל.
	3.	ניהול חכם של תכונות חסרות: ל-XGBoost יש מנגנון פנימי לטיפול בערכים חסרים. ב-Bagging אין מנגנון ייעודי, אלא משתמשים בעץ בסיסי שכלל לא בהכרח יודע “להחליט” על חסרים בדרך מתוחכמת.

ד. דמיון ושוני מול XGBoost
	•	דמיון:
	•	שניהם שייכים לעולם ה-Ensemble (שילוב מודלים רבים במטרה לשפר דיוק).
	•	שניהם לרוב עובדים עם עצי החלטה כמודלים בסיסיים (אם כי Bagging יכול גם לעטוף מודלים אחרים).
	•	שוני:
	1.	תהליך הלמידה: Bagging דוגם במקביל וממוצע; XGBoost בונה סדרה של עצים שכל אחד “מתקן” את שגיאות הקודמים.
	2.	רגישות ל-Overfitting: Bagging מפחית Variance, אבל לא תמיד מגביל את המורכבות של כל עץ. ב-XGBoost יש רגולריזציה מפורשת וקצב למידה (Learning Rate).
	3.	שימוש בדאטה: Bagging משתמש ב-Sample With Replacement (Bootstrap) לכל מודל בסיסי, בעוד XGBoost יכול להשתמש ב-Subsample עם או בלי Replacement, אך חשוב מכך – הוא משתמש במידע השגיאה באופן אדפטיבי.

⸻

4) Random Forest

א. סיכום קצר של השיטה

מה זה?
Random Forest הוא שכלול של Bagging עבור עצי החלטה, שבו לכל עץ נעשה גם Feature Subsampling (כלומר, בכל פיצול העץ בוחר תת-קבוצה רנדומלית של תכונות) בנוסף לדגימת Bootstrap של הדוגמאות. כך העצים עצמם הופכים לפחות מתואמים זה לזה, והתוצר הסופי (ממוצע/הצבעה) לרוב מדויק יותר.

ב. מדוע המודל הזה טוב (יתרונותיו)
	1.	Variety גבוהה של עצים: על ידי הגבלת סט התכונות בכל פיצול, מקטינים את המתאם בין העצים, ומשיגים אנסמבל חזק יותר ביחס ל-Bagging “רגיל”.
	2.	פשוט וכמעט תמיד עובד טוב: Random Forest מוצלח על מגוון רחב של דאטה, לעתים קרובות “ברירת מחדל” חזקה ב-ML.
	3.	אמידת חשיבות תכונות: מאפשר לחשב חשיבות (Importances) על בסיס כמה פעמים תכונה נבחרה לפיצול ועוד.

ג. יתרונות של XGBoost על פני Random Forest
	1.	Boosting במקום Bagging: XGBoost לא בונה עצים במקביל על Bootstrap, אלא בונה אותם ברצף, כאשר כל עץ מתמקד בתיקון שגיאות של הקודמים. לעיתים קרובות זה מוביל לשיפור דיוק בהשוואה ל-RF, במיוחד על דאטה מובנה (Tabular) בגדלים מסוימים.
	2.	רגולריזציה מפורשת: XGBoost כולל פרמטרים כמו \gamma, \lambda, ו-\alpha לעונשים, וכלי תיחום נוספים (למשל, min_child_weight, subsample, colsample_bytree). ב-Random Forest השליטה על Overfitting מגיעה דרך עומק העצים ושברים של תכונות, אך אין מנגנון “משוכלל” כמו ב-XGBoost.
	3.	טיפול בערכים חסרים: XGBoost מציע אסטרטגיות אוטומטיות לטיפול ב-Missing Values, בעוד ברוב המימושים הסטנדרטיים של RF יש צורך בגישות חיצוניות (אימפוטציה) או קבלת “NA” כערך מובחן.

ד. דמיון ושוני מול XGBoost
	•	דמיון:
	•	שניהם אלגוריתמי Ensemble מבוססי עצים.
	•	שניהם תומכים ב-Column Subsampling (ב-XGBoost: colsample_bytree) להפחתת מתאם בין עצים או בין סבבי Boosting.
	•	שוני:
	1.	מנגנון Ensemble: RF זה Bagging + Feature Subsampling; XGBoost זה Boosting אדפטיבי.
	2.	תהליך בנייה: ב-RF כל עץ נבנה במלואו בלי קשר לעצים אחרים (מלבד דגימה שונה). ב-XGBoost כל עץ תלוי בשגיאות/גרדיאנטים של הקודם.
	3.	רגולריזציה: ב-RF אין ענישה מפורשת על מורכבות העץ (מעבר להגדרת עומק מקסימלי וכו’), בעוד XGBoost מבצע זאת מפורשות.

⸻

5) Boosting

א. סיכום קצר של השיטה

מה זה?
Boosting, בהגדרתו הכללית, הוא תהליך שבו מריצים סדרת מודלים “חלשים” (למשל עצים קטנים) בזה אחר זה, וכל מודל חדש מנסה להתמקד בשגיאות שהמודל הקודם עשה.
	•	דוגמאות קלאסיות: AdaBoost, Gradient Boosting (GBM), וכן XGBoost שהוא שיפור על GBM.

ב. מדוע המודל הזה טוב (יתרונותיו)
	1.	למידה אדפטיבית: בכל שלב משקיעים מאמץ בעיקר בדוגמאות שעליהן המודל הקודם נכשל.
	2.	שיפור הדרגתי: בדומה ל-Gradient Descent, בכל שלב נצבר שיפור בפונקציית המטרה, עד שמקבלים מודל חזק ממודלים חלשים רבים.
	3.	גמישות: ניתן ליישם Boosting על סוגים שונים של מודלים בסיסיים (לא רק עצי החלטה), ועל מגוון פונקציות אובדן (רגרסיה, סיווג, ranking וכו’).

ג. יתרונות של XGBoost על פני Boosting “רגיל”

מאחר ש־XGBoost הוא סוג ספציפי של Gradient Boosting, היתרונות טמונים באופטימיזציות ובתוספות הרגולריזציה שהוא מציע:
	1.	רגולריזציה מובנית: פרמטרים \lambda, \alpha, \gamma, שלא בהכרח קיימים בכל מימוש GBM רגיל (למשל, ב-GBM הסטנדרטי ב־sklearn יש רגולריזציה מצומצמת יותר).
	2.	טכניקות האצה (Parallelization, Histogram): XGBoost ממומש ב־C++ וכולל אופטימיזציות לבניית עצים, תמיכה ב-GPU וכו’, מה שהופך אותו למהיר מאוד בקנה מידה גדול.
	3.	טיפול בערכים חסרים: XGBoost מציע “auto-learn” לנושא הערכים החסרים, בעוד ביישומים אחרים של Boosting עלולים לדרוש טיפול ידני.

ד. דמיון ושוני מול XGBoost
	•	דמיון:
	•	XGBoost הוא Gradient Boosting לכל דבר – בונה עצים בסדרת Boosting, משתמש בנגזרת של פונקציית אובדן, מעדכן תחזית שלב-אחר-שלב.
	•	שוני:
	1.	שיפורים טכניים: XGBoost הוא מימוש משוכלל ויעיל במיוחד של Boosting, עם מנגנוני Regularization והאצות.
	2.	יכולת טיפול בחסרים ופרמטרים ספציפיים: XGBoost כולל \gamma להענשת פיצולים, ויכולות colsample_bytree, colsample_bylevel וכו’, שלא תמיד מופיעים בכל מימוש Boosting.

⸻

סיכום
	1.	Dimensionality Reduction
	•	הכי שונה מ-XGBoost. זו טכניקת פרה-פרוססינג להקטנת מספר התכונות. לא מודל חיזוי בפני עצמו.
	•	XGBoost מספק חיזוי ישיר עם רגולריזציה, ללא “צמצום ממדים” מפורש (אלא אם משלבים בו preprocessing חיצוני).
	2.	Bootstrapping
	•	שיטת דגימה סטטיסטית המשמשת להערכת פרמטרים וסטיית תקן או בסיס ל-Ensemble (Bagging).
	•	XGBoost הוא אלגוריתם Boosting שלם, לא רק דגימה ב-Stochastic Manner.
	3.	Bagging
	•	אנסמבל שנבנה באמצעות דגימות Bootstrap ומיצוע תוצאות המודלים. מצמצם שונות, לא בהכרח מפחית Bias.
	•	XGBoost יותר אדפטיבי, בונה עצים עוקבים המתמקדים בשגיאות השלבים הקודמים. כולל מנגנוני רגולריזציה מפורטים.
	4.	Random Forest
	•	שיפור של Bagging על עצים, מוסיף Feature Subsampling בכל פיצול. טוב כמעט בכל מצב, קל לשימוש.
	•	XGBoost, כנציג Boosting, לרוב ישיג דיוק גבוה יותר בבעיות טבלאיות, ובעל שליטה מתקדמת על מורכבות העצים.
	5.	Boosting
	•	הכי קרוב ל-XGBoost. למעשה XGBoost הוא Gradient Boosting משופר.
	•	ההבדל הוא בעיקר באופטימיזציות (ביצועים, רגולריזציה, טיפול בחסרים) שהופכות את XGBoost לגרסה “משודרגת” במיוחד של Boosting.

כך ניתן לראות ש־XGBoost חולק הרבה מה-DNA עם Boosting (שכן הוא מימוש ספציפי) אבל שונה מהותית משיטות כמו Dimensionality Reduction (טכניקה להקטנת ממדים) או Bootstrapping (טכניקה סטטיסטית לדגימה). Bagging ו-Random Forest קרובים יותר, כי גם הם מודלים Ensemble מבוססי עצים, אבל רציונל האימון שלהם שונה (בניגוד לסדרת שיפורים הדרגתית, הם מאמנים עצים במקביל על מדגמים/תכונות רנדומליים).
