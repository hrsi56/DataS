### <code>plt.text()</code>

<br></br>
Introducing <code>plt.text()</code> to help with the next exercise
<br>
<br>
<code>plt.text()</code> helps to add lables to data points
<br>
<br>
Syntax:
<br>
<code>plt.text(x=x coordinate, y=y coordinate, s=string to be displayed)</code>
<br></br>

Let's start by importing `Matplotlib` and `Matplotlib.pyplot` as follows:

````python
import matplotlib.pyplot as plt
````

````python
# Example
a = [1,10, 20]
b = [3, 30, 10]

plt.plot(a,b)

plt.text(10,21, 'starts \nto \ndecrease -->')

plt.show()
````

<br></br>
x and y values in `plt.text(x, y, label)`:

    Since 'a' list elements are type 'integer', we can specify the value 10 for x-axis.
```python
    plt.text(10,21, 'starts \nto \ndecrease -->') # Data stored as type int
```
<br></br>

    If the elements were stored as type 'string', we would need to specify x as the index position.
    Eg index 1.
```python
    plt.text(1,21, 'starts \nto \ndecrease -->') # Data stored as type string
```

<br></br>

````python
# Example
a = ['1','10', '20']
b = [3, 30, 10]

plt.plot(a,b)

plt.text(1,21, 'starts \nto \ndecrease -->')

plt.show()
````
