### **אמידה נקודתית בשיטת מקסימום סבירות (Maximum Likelihood Estimation - MLE)**

---

### **מבוא:**

אמידת מקסימום סבירות (MLE) היא שיטה לאמידת פרמטרים של מודל סטטיסטי על ידי בחירת הערך שממקסם את פונקציית הסבירות (Likelihood Function).

---

### **דוגמה: התפלגות ברנולי (Bernoulli Distribution)**

נניח שיש לנו 4 דגימות מהתפלגות ברנולי:

\( $ x_1 = 1, \quad x_2 = 1, \quad x_3 = 0, \quad x_4 = 1 $\)

אנחנו רוצים לאמוד את \($p$\), שהוא ההסתברות להצלחה.

### **ייבוא ספריות:**

```python
import numpy as np
import matplotlib.pyplot as plt
```

### **יצירת פונקציית הסבירות עבור התפלגות ברנולי:**

```python
# Data: 4 samples from Bernoulli distribution
data = np.array([1, 1, 0, 1])
n = len(data)

# Likelihood function for Bernoulli
def likelihood_bernoulli(p, data):
    return np.prod(p**data * (1-p)**(1-data))

# p values for plotting
p_values = np.linspace(0, 1, 100)
likelihoods = [likelihood_bernoulli(p, data) for p in p_values]
```

### **מקסום פונקציית הסבירות ומציאת אומדן MLE:**

```python
# Finding the p that maximizes the likelihood
max_likelihood_index = np.argmax(likelihoods)
p_mle = p_values[max_likelihood_index]
```

### **הצגת פונקציית הסבירות:**

```python
plt.figure(figsize=(10, 6))
plt.plot(p_values, likelihoods, label='Likelihood Function', color='blue')
plt.axvline(p_mle, color='red', linestyle='--', label=f'MLE Estimate (p = {p_mle:.2f})')
plt.xlabel('p')
plt.ylabel('Likelihood')
plt.title('Maximum Likelihood Estimation for Bernoulli Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

### **תוצאות:**

```python
print(f'MLE Estimate for p: {p_mle:.4f}')
```

### **אמידת פרמטר בהתפלגות פואסון (Poisson Distribution):**

נניח שיש לנו דגימות מהתפלגות פואסון:

\($ x_1 = 3, \quad x_2 = 0, \quad x_3 = 2, \quad x_4 = 7$ \)

אנחנו רוצים לאמוד את \( $\lambda $\), שהוא מספר האירועים הממוצע בזמן נתון.

```python
import numpy as np
from scipy.special import factorial  # Import factorial from scipy

# Data: 4 samples from Poisson distribution
data_poisson = np.array([3, 0, 2, 7])

# Likelihood function for Poisson
def likelihood_poisson(lmbda, data):
    return np.prod((np.exp(-lmbda) * lmbda**data) / factorial(data))

# λ values for plotting
lambda_values = np.linspace(0.1, 10, 100)
likelihoods_poisson = [likelihood_poisson(lmbda, data_poisson) for lmbda in lambda_values]
```

### **מקסום פונקציית הסבירות עבור פואסון:**

```python
# Finding the λ that maximizes the likelihood
max_likelihood_index_poisson = np.argmax(likelihoods_poisson)
lambda_mle = lambda_values[max_likelihood_index_poisson]
```

### **הצגת פונקציית הסבירות עבור פואסון:**

```python
plt.figure(figsize=(10, 6))
plt.plot(lambda_values, likelihoods_poisson, label='Likelihood Function', color='blue')
plt.axvline(lambda_mle, color='red', linestyle='--', label=f'MLE Estimate (λ = {lambda_mle:.2f})')
plt.xlabel('λ')
plt.ylabel('Likelihood')
plt.title('Maximum Likelihood Estimation for Poisson Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

### **אמידת פרמטר בהתפלגות מעריכית (Exponential Distribution):**

נניח שיש לנו דגימות מהתפלגות מעריכית:

\($ x_1 = 3.22, \quad x_2 = 1.81, \quad x_3 = 2.47$ \)

אנחנו רוצים לאמוד את \( $\theta$ \), שהוא הקצב (rate parameter) של ההתפלגות.

```python
# Data: 3 samples from Exponential distribution
data_exponential = np.array([3.22, 1.81, 2.47])

# Likelihood function for Exponential
def likelihood_exponential(theta, data):
    return np.prod(theta * np.exp(-theta * data))

# θ values for plotting
theta_values = np.linspace(0.1, 2, 100)
likelihoods_exponential = [likelihood_exponential(theta, data_exponential) for theta in theta_values]
```

### **מקסום פונקציית הסבירות עבור מעריכית:**

```python
# Finding the θ that maximizes the likelihood
max_likelihood_index_exponential = np.argmax(likelihoods_exponential)
theta_mle = theta_values[max_likelihood_index_exponential]
```

### **הצגת פונקציית הסבירות עבור מעריכית:**

```python
plt.figure(figsize=(10, 6))
plt.plot(theta_values, likelihoods_exponential, label='Likelihood Function', color='blue')
plt.axvline(theta_mle, color='red', linestyle='--', label=f'MLE Estimate (θ = {theta_mle:.2f})')
plt.xlabel('θ')
plt.ylabel('Likelihood')
plt.title('Maximum Likelihood Estimation for Exponential Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

### **תוצאות עבור פואסון ומעריכית:**

```python
print(f'MLE Estimate for λ (Poisson): {lambda_mle:.4f}')
print(f'MLE Estimate for θ (Exponential): {theta_mle:.4f}')
```
