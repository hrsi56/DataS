## Exercise 16:
Plot a histogram for 3 different data sets.
<br>`data1 = np.random.normal(0, 0.8, 1000)`
<br>`data2 = np.random.normal(-2, 1, 1000)`
<br>`data3 = np.random.normal(3, 2, 1000)`

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
data1 = np.random.normal(0, 0.8, 1000)
data2 = np.random.normal(-2, 1, 1000)
data3 = np.random.normal(3, 2, 1000)

# Define consistent bins
bins = 100  # Number of bins

# Compute histogram data for consistent y-axis
_, bin_edges = np.histogram(np.concatenate([data1, data2, data3]), bins=bins)
counts1, _ = np.histogram(data1, bins=bin_edges)
counts2, _ = np.histogram(data2, bins=bin_edges)
counts3, _ = np.histogram(data3, bins=bin_edges)
counts_combined, _ = np.histogram(np.concatenate([data1, data2, data3]), bins=bin_edges)

# Find the maximum y-limit across all histograms
max_y = max(counts1.max(), counts2.max(), counts3.max(), counts_combined.max())

# Plot separate histograms
plt.figure(figsize=(10, 6))
plt.hist(data1, bins=bin_edges, alpha=0.5, label='Data 1', color='blue', edgecolor='black')
plt.hist(data2, bins=bin_edges, alpha=0.5, label='Data 2', color='green', edgecolor='black')
plt.hist(data3, bins=bin_edges, alpha=0.5, label='Data 3', color='orange', edgecolor='black')
plt.ylim(0, max_y + 10)  # Set consistent y-axis limit

plt.title('Separate Histograms with Consistent Binning and Y-Axis', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.3, linestyle='--')
plt.show()

# Plot combined histogram
plt.figure(figsize=(10, 6))
plt.hist([data1, data2, data3], bins=bin_edges, alpha=0.7, label=['Data 1', 'Data 2', 'Data 3'], edgecolor='black')
plt.ylim(0, max_y + 10)  # Set consistent y-axis limit

plt.title('Combined Histogram with Consistent Binning and Y-Axis', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.3, linestyle='--')
plt.show()
```
