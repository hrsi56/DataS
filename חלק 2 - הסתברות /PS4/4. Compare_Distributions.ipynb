### **השוואה בין שתי התפלגויות באמצעות בדיקת השערות דו-צדדית**

---

### **מבוא:**

בבדיקה זו נשווה בין שתי התפלגויות בינומיאליות שונות ונבחן האם יש הבדל מובהק סטטיסטית ביניהן. נשתמש בשתי שיטות לבדיקה:

1. **בדיקה באמצעות ערך קריטי (Critical Value).**  
2. **בדיקה באמצעות p-value.**

### **הגדרת הבעיה:**

ניקח שתי התפלגויות בינומיאליות שונות:

- **התפלגות 1:**  
  \( $\text{Binomial}(n=10, p=0.2)$\)  
  \($\mu_1 = 10 \cdot 0.2 = 2$\)

- **התפלגות 2:**  
  \(\ $ text{Binomial}(n=10, p=0.4) $ \)  
  \( $ \mu_2 = 10 \cdot 0.4 = 4 $ \)

### **השערות לבדיקה דו-צדדית:**

- **\($H_0$: $\mu_1$ = $\mu_2$\)** – אין הבדל בין הממוצעים.
- **\($H_1$: $\mu_1 \neq \mu_2 $\)** – יש הבדל בין הממוצעים.

### **ייבוא ספריות:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

### **הגדרת פרמטרים תאורטיים:**

```python
n_trials = 10
p1 = 0.2  # הסתברות הצלחה בהתפלגות 1
p2 = 0.4  # הסתברות הצלחה בהתפלגות 2
sample_size = 100
num_iterations = 1000

# תוחלות וסטיות תקן תאורטיות
mu1 = n_trials * p1
mu2 = n_trials * p2
std1 = np.sqrt(n_trials * p1 * (1 - p1))
std2 = np.sqrt(n_trials * p2 * (1 - p2))

# שגיאת התקן של ההפרש בין הממוצעים
standard_error_diff = np.sqrt((std1**2 / sample_size) + (std2**2 / sample_size))
```

### **בדיקה דו-צדדית באמצעות ערך קריטי ו-p-value:**

```python
# רמת מובהקות לבדיקה דו-צדדית
alpha = 0.05
z_critical_two_tailed = norm.ppf(1 - alpha / 2)

# דחיות בשתי השיטות
rejections_critical = 0
rejections_p_value = 0

# בדיקות על 1000 דגימות
for _ in range(num_iterations):
    sample1 = np.random.binomial(n=n_trials, p=p1, size=sample_size)
    sample2 = np.random.binomial(n=n_trials, p=p2, size=sample_size)
    
    mean_diff = np.mean(sample1) - np.mean(sample2)
    
    # חישוב סטטיסטי הבדיקה Z
    z_score = (mean_diff) / standard_error_diff
    
    # בדיקה באמצעות ערך קריטי
    if abs(z_score) > z_critical_two_tailed:
        rejections_critical += 1
    
    # חישוב p-value לבדיקה דו-צדדית
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    if p_value < alpha:
        rejections_p_value += 1
```

### **הצגת גרפים להשוואת ההתפלגויות:**

```python
plt.figure(figsize=(12, 6))

# התפלגות 1
plt.hist(sample1, bins=range(0, n_trials+2), alpha=0.6, label='התפלגות 1 (p=0.2)', edgecolor='black')
plt.axvline(np.mean(sample1), color='blue', linestyle='dashed', linewidth=2, label=f'ממוצע התפלגות 1: {np.mean(sample1):.2f}')

# התפלגות 2
plt.hist(sample2, bins=range(0, n_trials+2), alpha=0.6, label='התפלגות 2 (p=0.4)', edgecolor='black')
plt.axvline(np.mean(sample2), color='red', linestyle='dashed', linewidth=2, label=f'ממוצע התפלגות 2: {np.mean(sample2):.2f}')

plt.xlabel('מספר הצלחות')
plt.ylabel('תדירות')
plt.title('השוואת שתי התפלגויות בינומיאליות')
plt.legend()
plt.grid(True)
plt.show()
```

### **סיכום תוצאות הבדיקה:**

```python
print(f"מספר דחיות של H0 באמצעות ערך קריטי: {rejections_critical} מתוך {num_iterations} איטרציות")
print(f"מספר דחיות של H0 באמצעות p-value: {rejections_p_value} מתוך {num_iterations} איטרציות")

print(f"\nאחוז דחיות באמצעות ערך קריטי: {(rejections_critical / num_iterations) * 100:.2f}%")
print(f"אחוז דחיות באמצעות p-value: {(rejections_p_value / num_iterations) * 100:.2f}%")
```
