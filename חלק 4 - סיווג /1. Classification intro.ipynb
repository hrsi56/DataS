# י פונקציית החלטה ($Decision\ Function$) י  

## י הגדרה בסיסית: י  

י פונקציית החלטה, או $Decision\ Function$, היא פונקציה שלוקחת וקטור של תכונות ($features$) כקלט ומחזירה ערך מספרי, אשר לפיו מחליטים לאיזו קבוצה ($class$) התצפית שייכת. י  

י הפונקציה לרוב תהיה מהצורה: י  

$$
f(x) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_p x_p
$$

י כאשר: י  
- י $x = (x_1, x_2, \dots, x_p)$ הוא וקטור התכונות. י  
- י $\beta = (\beta_0, \beta_1, \dots, \beta_p)$ הם מקדמי המודל שנלמדו מהנתונים. י  

י הערך של $f(x)$ משמש להחלטה לאיזו קבוצה התצפית תסווג: י  
- י אם $f(x)$ גדול מסף מסוים ($threshold$), התצפית תסווג לקבוצה אחת. י  
- י אם קטן מהסף, היא תסווג לקבוצה השנייה. י  

י בד"כ, בסיווג בינארי (שתי קבוצות), הסף הוא $0$. י  

---

## י דוגמה ויזואלית ברורה: י  

י למשל, במודל סיווג בינארי לינארי, פונקציית ההחלטה היא קו ישר (או היפר-מישור בממדים גבוהים יותר), שמחלק את המרחב לשני חלקים. י  

י פונקציית ההחלטה במקרה זה תהיה: י  

$$
f(x) = \beta_0 + \beta_1 x_1 + \beta_2 x_2
$$

י והסיווג מתבצע כך: י  
- י אם $f(x) > 0$, התצפית שייכת לקבוצה אחת (לדוגמה, $Class\ 1$). י  
- י אם $f(x) < 0$, התצפית שייכת לקבוצה אחרת (לדוגמה, $Class\ 0$). י  

---

## י דוגמה ויזואלית ב-$Python$ (עם $Decision\ Function$): י  

י נוסיף על הגרף את הערכים של הפונקציה $f(x)$: י  
- י נקודות על גבול ההחלטה ייתנו $f(x) = 0$. י  
- י נקודות מעל הגבול ייתנו $f(x) > 0$, ומתחתיו – $f(x) < 0$. י  

י נציג את הפונקציה והגבולות על הגרף הקודם להמחשה: י  

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

# Generate synthetic classification data
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0,
                           n_informative=2, n_clusters_per_class=1, random_state=42)

# Fit a Logistic Regression classifier
clf = LogisticRegression()
clf.fit(X, y)

# Plot the points and decision boundary
plt.figure(figsize=(10, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label="Class 0", edgecolor='k')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label="Class 1", edgecolor='k')

# Decision boundary
x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()

xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, levels=np.linspace(0, 1, 3))

# Labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Classification Example with Logistic Regression')
plt.legend()
plt.grid(True)
plt.show()
# Plotting Decision Function explicitly

# Calculate Decision Function values over the grid
Z_decision = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z_decision = Z_decision.reshape(xx.shape)

# Plot data points
plt.figure(figsize=(10, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label="Class 0", edgecolor='k')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label="Class 1", edgecolor='k')

# Plot decision boundary and decision function levels
plt.contour(xx, yy, Z_decision, levels=[0], linewidths=2, colors='black')
contour = plt.contourf(xx, yy, Z_decision, alpha=0.3, levels=20, cmap='RdBu_r')
plt.colorbar(contour, label='Decision Function value')

# Labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Function Visualization (Logistic Regression)')
plt.legend()
plt.grid(True)
plt.show()
```


הסבר ויזואלי:

הגרף מציג בצורה מפורשת את פונקציית ההחלטה (Decision Function):  
	•	הקו השחור העבה (Decision Boundary) הוא המקום שבו f(x)=0.  
	•	האזורים האדומים והכחולים מציגים את הערכים של פונקציית ההחלטה עצמה:  
	•	האזורים האדומים מציינים ערכים חיוביים (התצפית מסווגת ל-Class 1).  
	•	האזורים הכחולים מציינים ערכים שליליים (התצפית מסווגת ל-Class 0).  

ככל שהתצפית רחוקה יותר מגבול ההחלטה, כך הביטחון (או הערך המוחלט של פונקציית ההחלטה) גדול יותר.

## י מטריצת בלבול ($Confusion Matrix$) י  

י תהליך הסיווג מוערך באמצעות מטריצת בלבול ($Confusion Matrix$). י  

י המטריצה נראית כך (בדוגמה של סיווג בינארי): י  

|                    | י ניבוי חיובי י   | י ניבוי שלילי י  |
|--------------------|-------------------|-------------------|
| י חיובי אמיתי י    | $TP$ ($True Positive$) | $FN$ ($False Negative$) |
| י שלילי אמיתי י    | $FP$ (False Positive) | $TN$ (True Negative) |

| י סימון י     | י משמעות י                          |
|---------------|-----------------------------------|
| $TP$          | $True\ Positive$                  |
| $FN$          | $False\ Negative$                 |
| $FP$          | $False\ Positive$                 |
| $TN$          | $True\ Negative$                  |

⸻

## י מדדים מרכזיים ממטריצת הבלבול: י  
• י $Recall$ – כמה מתוך החיוביים זוהו ($TP/(TP+FN)$). י  
• י $Specificity$: אחוז השליליים שזוהו נכון: $\frac{TN}{TN + FP}$ י  
• י $Precision$: מתוך אלו שחזו חיוביים, כמה באמת חיוביים: $\frac{TP}{TP + FP}$ י  

י לדוגמה: י  
• י בבדיקת מחלה קשה, חשוב שיהיה לנו $Recall$ גבוה, גם אם ה-$Precision$ יורד מעט (עדיף לזהות כמה שיותר מקרים חיוביים, גם במחיר טעויות). י  
• י בהחלטה על מתן הלוואה, חשוב דווקא דיוק חיובי ($Precision$ גבוה), גם אם פחות מזהים מקרים גבוליים כחיוביים. י
