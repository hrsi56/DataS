# 📊 סיכום: רגרסיה ליניארית פשוטה (Simple Linear Regression)

## מבוא
רגרסיה ליניארית פשוטה מניחה כי קיים קשר **ליניארי** בין משתנה תלוי $ Y $ לבין משתנה בלתי תלוי אחד $ X $.  י
המודל מתואר על ידי המשוואה:

$$[ Y = \beta_0 + \beta_1 X + \varepsilon ]$$

כאשר:
- $ \beta_0 $ – חותך (intercept)
- $ \beta_1 $ – מקדם הרגרסיה (slope)
- $ \varepsilon $ – שגיאה מקרית עם תוחלת 0

**מטרת הרגרסיה** היא למצוא את הערכים המוערכים $ \hat{\beta}_0, \hat{\beta}_1 $ כך שנוכל לחזות את $ Y $ עבור ערך נתון של $ X $. י




## שגיאות, חריגים ומדדים ברגרסיה ליניארית

### סטיות התקן של מקדמי הרגרסיה (Standard Errors - SE)
סטיית התקן של מקדמי הרגרסיה משמשת להערכת חוסר הוודאות של ההערכות שלהם. ככל שהסטייה התקנית קטנה יותר, כך האומדן של הפרמטרים במודל מדויק יותר.

#### **סטיית התקן של $\hat{\beta}_0$ (החותך)**
$$
SE(\hat{\beta}_0)^2 = \sigma^2 \left[ \frac{1}{n} + \frac{\bar{X}}{\sum_{i=1}^{n} (X_i - \bar{X})^2} \right]
$$
כאשר:
- $ \sigma^2 $ – שונות השגיאות במודל.
- $ n $ – מספר התצפיות.
- $ \bar{X} $ – ממוצע המשתנה הבלתי תלוי $X$.

#### **סטיית התקן של $\hat{\beta}_1$ (השיפוע)**
$$ 
SE(\hat{\beta}_1)^2 = \frac{\sigma^2}{\sum_{i=1}^{n} (X_i - \bar{X})^2}
$$ 
- ככל שהפיזור של ערכי $X$ גבוה יותר, כך השגיאה באמידת $\hat{\beta}_1$ קטנה יותר.
- כאשר המשתנים הבלתי תלויים מאוד קרובים אחד לשני (נמוכה שונות ב-$X$), השגיאה באמידה תהיה גבוהה יותר.

---

## אומדן השונות של השגיאות
$$
\hat{\sigma}^2 = \frac{RSS}{n-2} = \frac{\sum_{i=1}^{n} e_i^2}{n-2}
$$
כאשר:
- $RSS$ - סכום ריבועי השאריות, כלומר כמה רחוקות התחזיות מהערכים האמיתיים.
- $e_i$ - השאריות של המודל עבור כל תצפית.
- $n-2$ - דרגות החופש של המודל (הורדנו שני פרמטרים $\beta_0, \beta_1$).

---

## **מדד $R^2$ - אחוז השונות שהוסברה על ידי המודל**
$$
R^2 = 1 - \frac{RSS}{TSS}
$$
כאשר:
$$
TSS = \sum_{i=1}^{n} (y_i - \bar{y})^2
$$
$$RSS = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

- **$TSS$ (Total Sum of Squares)** מודד את השונות הכוללת במשתנה התלוי $Y$ סביב ממוצע המדגם.
- **$RSS$ (Residual Sum of Squares)** מודד את השונות הלא מוסברת על ידי המודל, כלומר סכום ריבועי השאריות.
- י$R^2$ מודד את אחוז השונות ב-$Y$ המוסברת על ידי $X$, כך ש:
  - $R^2 = 1$ מציין התאמה מלאה.
  - $R^2 = 0$ מציין שהמודל אינו מסביר כלל את השונות ב-$Y$ י.


- י $R^2$ מודד את אחוז השונות המוסברת:
  $$
  R^2 = \frac{TSS - RSS}{TSS} = 1 - \frac{RSS}{TSS}
  $$
  ככל שהמודל מתאים טוב יותר, כך $RSS$ קטן יותר ו-$R^2$ קרוב ל-1.

---

## **השוואה בין $ SE $ ל- $SD$**
| **מאפיין** | **Standard Error (SE)** | **Standard Deviation (SD)** |
|------------|------------------------|----------------------------|
| **משמעות** | מודד חוסר ודאות בהערכת מקדם (למשל, $\beta$) | מודד פיזור הנתונים סביב הממוצע |
| **נוסחה כללית** | $ SE = \frac{SD}{\sqrt{n}} $ | $ SD = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n-1}} $ |
| **משמש עבור** | הערכת שגיאה של ממוצע או מקדמים | מדידת הפיזור של משתנה |
| **מה משפיע עליו?** | גודל המדגם – קטן יותר ב-$n$ גדול | שונות המשתנה |
| **פרשנות** | ככל ש-$ SE $ קטן, כך האומדן (למשל, $ \hat{\beta} $) מדויק יותר | ככל ש-$ SD $ גדול יותר, כך הפיזור רחב יותר |

#### **ההבדל המרכזי בין SE ל-SD**
- **$SD$** מודד את הפיזור של הנתונים המקוריים סביב הממוצע שלהם.
- **$SE$** מודד את חוסר הוודאות באומדן של פרמטרים (כגון ממוצע או מקדמים).

📌 **דוגמה להמחשה:**  
- אם נמדוד את הגובה של 100 אנשים, **$SD$** ימדוד כמה הנתונים רחוקים מהממוצע של הגובה.
- **$SE$** ימדוד כמה ההערכה של הממוצע לא בטוחה – ככל שגודל המדגם גדול יותר, כך $SE$ קטן יותר.

---

## **📌 סיכום**
✔ **$SE$** מודד את חוסר הוודאות באומדן מקדמי הרגרסיה.  
✔ **$R^2$** מודד את התאמת המודל לנתונים.  
✔ **$SD$ ו-$SE$** מודדים פיזור, אך $SE$ מושפע גם מגודל המדגם.  
✔ **$RSS$** מודד את השונות שלא הוסברה ע"י המודל, בעוד **$TSS$** מודד את כלל השונות בנתונים.  

📌 **לניתוח נתונים מדויק, יש לבדוק גם את סטיות התקן וגם את איכות ההתאמה של המודל! 🚀**
 

## אמידת המקדמים – שיטת הריבועים הפחותים (Least Squares Estimation)
כדי לאמוד את הפרמטרים $ \beta_0 $ ו-$ \beta_1 $, אנו מחפשים ערכים הממזערים את **סכום ריבועי השאריות** ($RSS$):

$$RSS = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

**אומדני המקדמים של הרגרסיה מתקבלים מהנוסחאות הבאות:**

$$\hat{\beta}_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}$$

$$\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x}$$

#### **הסבר לנוסחאות:**
- י$ \hat{\beta}_1 $ מייצג את **שיפוע קו הרגרסיה**, כלומר את קצב השינוי של $ Y $ לכל שינוי יחידתי ב- $ X $. י  
- י$ \hat{\beta}_0 $ הוא **החותך (Intercept)**, כלומר ערך התחזית של $ Y $ כאשר $ X = 0 $.י
- י$י \bar{x} $ ו- $ \bar{y} $ הם ממוצעי המשתנים הבלתי תלויים והתלויים בהתאמה.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ISLP import load_data
import statsmodels.api as sm

# Load the ISLP Auto dataset
Auto_set = load_data('Auto')

# Independent variable
X = sm.add_constant(Auto_set['horsepower'])

# Dependent variable
y = Auto_set['mpg']

# Fit the model
model = sm.OLS(y, X).fit()

# Add predictions to the dataset
Auto_set['predicted_mpg'] = model.predict(X)

# Ensure compatibility by converting to NumPy arrays
horsepower = Auto_set['horsepower'].to_numpy()
predicted_mpg = Auto_set['predicted_mpg'].to_numpy()

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot of actual data
plt.scatter(horsepower, y, alpha=0.7, label='Actual Data', color='blue')

# Plot of regression line
plt.plot(horsepower, predicted_mpg, color='red', label='Regression Line', linewidth=2)

# Add labels, title, and legend
plt.title('Horsepower vs MPG with OLS Regression Line', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('MPG', fontsize=14)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from ISLP import load_data

# Load the Auto dataset
Auto_set = load_data('Auto')

# Create the design matrix (X) with an intercept and 'horsepower' column
X = sm.add_constant(Auto_set['horsepower'])  # Adds an intercept term

# Target variable (y)
y = Auto_set['mpg']

# Fit the OLS model
model = sm.OLS(y, X)
results = model.fit()

# Print the summary of the results
print(results.summary())

# Generate predicted values for the range of horsepower
horsepower_range = np.linspace(Auto_set['horsepower'].min(), Auto_set['horsepower'].max(), 500)
X_range = sm.add_constant(horsepower_range)  # Add intercept to the range
predicted_mpg = results.predict(X_range)

# Plot the data points and the regression line
plt.figure(figsize=(10, 6))
plt.scatter(Auto_set['horsepower'], Auto_set['mpg'], alpha=0.7, label='Actual Data', color='blue')
plt.plot(horsepower_range, predicted_mpg, color='red', label='Regression Line', linewidth=2)
plt.title('Horsepower vs MPG with OLS Regression Line', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('MPG', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from ISLP import load_data

# Load the Auto dataset
Auto_set = load_data('Auto')

# Define the target variable (y)
y = Auto_set['mpg']

# Helper function to fit a regression model and return predicted values
def fit_and_predict(transformed_X, x_label):
    X = sm.add_constant(transformed_X)
    model = sm.OLS(y, X)
    results = model.fit()
    print(f"Regression Summary for {x_label}:")
    print(results.summary())
    return results

# Set up the horsepower range for plotting smooth lines
horsepower_range = np.linspace(Auto_set['horsepower'].min(), Auto_set['horsepower'].max(), 500)

# Original regression (linear relationship)
linear_model = fit_and_predict(Auto_set['horsepower'], 'horsepower')
linear_pred = linear_model.predict(sm.add_constant(horsepower_range))

# Regression with sqrt(horsepower)
sqrt_model = fit_and_predict(np.sqrt(Auto_set['horsepower']), 'sqrt(horsepower)')
sqrt_pred = sqrt_model.predict(sm.add_constant(np.sqrt(horsepower_range)))

# Regression with log(horsepower)
# Add a small constant to avoid log(0)
log_model = fit_and_predict(np.log(Auto_set['horsepower'] ) , 'log(horsepower)')
log_pred = log_model.predict(sm.add_constant(np.log(horsepower_range )))

# Regression with horsepower^2
squared_model = fit_and_predict(Auto_set['horsepower']**2, 'horsepower^2')
squared_pred = squared_model.predict(sm.add_constant(horsepower_range**2))

# Plotting all the regressions
plt.figure(figsize=(12, 8))

# Scatter plot of actual data
plt.scatter(Auto_set['horsepower'], Auto_set['mpg'], alpha=0.7, label='Actual Data', color='blue')

# Plot the regression lines
plt.plot(horsepower_range, linear_pred, label='Linear: mpg ~ horsepower', color='red', linewidth=2)
plt.plot(horsepower_range, sqrt_pred, label='Sqrt: mpg ~ sqrt(horsepower)', color='green', linewidth=2)
plt.plot(horsepower_range, log_pred, label='Log: mpg ~ log(horsepower)', color='orange', linewidth=2)
plt.plot(horsepower_range, squared_pred, label='Squared: mpg ~ horsepower^2', color='purple', linewidth=2)

# Final plot adjustments
plt.title('Horsepower vs MPG with Transformations', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('MPG', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
