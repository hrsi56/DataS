## י איך משתמשים ברגרסיה לינארית לסיווג? י  

י כאשר אנו רוצים להשתמש ברגרסיה לינארית לבעיה של סיווג בינארי (שתי קטגוריות בלבד), נבצע את התהליך הבא: י  
• י נקודד את הקטגוריה למספרים (לדוגמה $0$ ו-$1$). י  
• י נבנה מודל לינארי ונבצע את הסיווג לפי הכלל הבא: י  
  - י אם התוצאה של המודל $(\hat{y}) > 0.5$, נסווג כקטגוריה $1$. י  
  - י אם התוצאה $(\hat{y}) \leq 0.5$, נסווג כקטגוריה $0$. י  

---

## י למה זה הגיוני? י  

י באופן תיאורטי, אם המודל מחזיר ערך של $0.8$, ניתן לפרש זאת כהסתברות של $80\%$ לקבל את הערך $1$. י  

י עם זאת, ישנן בעיות בשיטה זו, מאחר והמודל הלינארי אינו מוגבל בין $0$ ל-$1$ ולכן עלול לנבא הסתברויות לא חוקיות. י  

---

## י דוגמה מספרית פשוטה י  

י נניח את הנתונים הבאים של שעות לימוד מול הצלחה במבחן: י  

| י שעות לימוד י | י האם עבר? י |
|---------------|---------------|
| $1$           | $0$           |
| $2$           | $0$           |
| $3$           | $0$           |
| $4$           | $1$           |
| $5$           | $1$           |
| $6$           | $1$           |

י נניח שמשוואת הרגרסיה הלינארית שהתקבלה היא: י  
$$
\hat{y} = -0.4 + 0.2 \cdot X
$$

י לדוגמה, עבור תלמיד שלמד $5.5$ שעות: י  
$$
\hat{y} = -0.4 + 0.2 \cdot 5.5 = 0.7
$$

י הערך גדול מ-$0.5$, ולכן המודל יסווג אותו כ"עבר". י  

י לעומת זאת, תלמיד שלמד $2$ שעות: י  
$$
\hat{y} = -0.4 + 0.2 \cdot 2 = 0
$$

י התוצאה קטנה מ-$0.5$, ולכן המודל יסווג אותו כ"לא עבר". י  

---

## י מושגי יסוד י  

### י משתנה תגובה ($Response\ Variable$ / $Target$ / $Label$): י  

י המשתנה שאותו רוצים לנבא, לדוגמה "עבר"/"לא עבר". י  

### י משתני ניבוי ($Predictors / Features$): י  

י משתנים שעל בסיסם נעשה הסיווג, למשל שעות לימוד. י  

### י מרחב התכונות ($Feature Space$): י  

י כל תצפית מיוצגת כנקודה במרחב רב-ממדי של תכונות, וסיווג פירושו חלוקת המרחב לאזורים לפי קטגוריות. י  

---

## י מטריצת בלבול ($Confusion Matrix$): י  

י מטריצה שמציגה את ביצועי המודל בצורה הבאה (דוגמה בינארית): י  

|                    | י ניבוי חיובי י     | י ניבוי שלילי י     |
|--------------------|---------------------|----------------------|
| י חיובי אמיתי י    | $TP$ ($True\ Positive$) | $FN$ ($False\ Negative$) |
| י שלילי אמיתי י    | $FP$ ($False Positive$)| $TN$ ($True Negative$) |

י הסבר המונחים: י  
- י $TP$: סיווג נכון של מקרה חיובי ($True Positive$). י  
- י $TN$: סיווג נכון של מקרה שלילי ($True Negative$). י  
- י $FP$: סיווג שגוי של מקרה שלילי כחיובי ($False Positive$). י  
- י $FN$: סיווג שגוי של מקרה חיובי כשלילי ($False Negative$). י  

---

## י מדדי ביצועים מרכזיים: י  

- י דיוק ($Accuracy$): י  
$$
\frac{TP + TN}{TP + FP + TN + FN}
$$  

- י רגישות ($Sensitivity / Recall$): אחוז החיוביים שזוהו נכון: י  
$$
\frac{TP}{TP + FN}
$$  

- י ספציפיות ($Specificity$): אחוז השליליים שזוהו נכון: י  
$$
\frac{TN}{TN + FP}
$$  

- י דיוק חיובי ($Precision$): מבין אלה שסווגו כחיוביים, כמה באמת חיוביים? י  
$$
\frac{TP}{TP + FP}
$$  

- י קצב הטעות ($Error\ Rate$): אחוז הטעויות הכללי: י  
$$
\frac{FP + FN}{TP + FP + TN + FN}
$$  

---

## י יתרונות וחסרונות של שימוש ברגרסיה לינארית לסיווג י  

י **יתרונות:** י  
- י פשוטה וקלה להסבר וליישום. י  

י **חסרונות:** י  
- י לא מוגבלת בין $0$ ל-$1$. י  
- י מושפעת מתצפיות חריגות ($Outliers$). י  
- י פחות מתאימה למקרים שבהם דרושה הערכה מדויקת של הסתברות. י  



## י דוגמה מעשית  פשוטה (תכונה 1) ב-$Python$: י  

י ניצור דוגמה קצרה להמחשת השימוש ברגרסיה לינארית עבור סיווג בינארי: י  
- י נבנה מודל רגרסיה לינארית פשוט, ונבחן כיצד הוא מסווג נתונים סינתטיים. י  


```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

# יצירת נתונים מלאכותיים לסיווג
X, y = make_classification(n_samples=100, n_features=1, n_informative=1,
                           n_redundant=0, n_clusters_per_class=1, random_state=42)

# אימון מודל רגרסיה לינארית
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# יצירת תחזית
y_pred_continuous = lin_reg.predict(X)

# סיווג לפי סף 0.5
y_pred_class = (y_pred_continuous > 0.5).astype(int)

# הצגת התוצאות
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='blue', label='Actual Class')
plt.scatter(X, y_pred_continuous, color='red', alpha=0.5, label='Predicted (Continuous)')
plt.axhline(0.5, color='black', linestyle='--', label='Decision Threshold (0.5)')
plt.xlabel('Feature (X)')
plt.ylabel('Class / Predicted Value')
plt.title('Linear Regression for Binary Classification')
plt.legend()
plt.grid(True)
plt.show()
```

## י מתי כדאי להשתמש בשיטה הזו? י  

י רגרסיה לינארית לסיווג יכולה להיות שימושית כשאנחנו רוצים פתרון מהיר ופשוט. עם זאת, בדרך כלל היא נחותה משיטות כמו רגרסיה לוגיסטית או $Linear\ Discriminant\ Analysis$ (שנלמד בהמשך). י  



## י דוגמה ויזואלית: סיווג בינארי בהתאם ל-$2$ תכונות באמצעות רגרסיה לינארית וגבול ההחלטה ($Decision\ Boundary$) י  

י בדוגמה הבאה אנו משתמשים בנתונים סינתטיים שיש להם $2$ תכונות בלבד, כדי להמחיש כיצד מתבצע סיווג בינארי באמצעות רגרסיה לינארית. י  

- י כל נקודה בגרף היא תצפית בעלת שתי תכונות, המסומנת לפי הקטגוריה האמיתית שלה. י  
- י הקו האדום בגרף הוא גבול ההחלטה של המודל, כלומר המקום שבו התצפית בדיוק נמצאת במרכז ההחלטה בין שתי הקבוצות ($\hat{y}(x) = 0.5$). י  

י כל תצפית שנמצאת מעל הקו האדום מסווגת על ידי המודל לקבוצה אחת, ותצפיות מתחתיו מסווגות לקבוצה השנייה. י  

⸻

י הקו הזה מייצג בדיוק את גבול ההחלטה ($Decision\ Boundary$). י  
י כלומר, הרגרסיה הלינארית "חוצה" את מרחב התכונות לשני אזורים מוגדרים על ידי קו ישר. י  

### י גבול ההחלטה ($Decision\ Boundary$) בדוגמה שלנו הוא הקו שבו הערך שהמודל מנבא הוא בדיוק $0.5$. י  

י במילים אחרות, זוהי הנקודה שבה המודל "מתלבט" בין שתי הקבוצות: י  

$$
\hat{y}(x) = 0.5
$$

י המודל שהשתמשנו בו הוא רגרסיה לינארית, כלומר הצורה הכללית של המודל היא: י  

$$
\hat{y}(x) = \beta_0 + \beta_1 x_1 + \beta_2 x_2
$$

י כאשר: י  
- י $x_1, x_2$ הם התכונות. י  
- י $\beta_0, \beta_1, \beta_2$ הם מקדמי המודל. י  

⸻

## י חישוב מדויק של גבול ההחלטה: י  

י גבול ההחלטה מתקבל על ידי הצבת $\hat{y}(x) = 0.5$: י  

$$
0.5 = \beta_0 + \beta_1 x_1 + \beta_2 x_2
$$

י נבודד את המשוואה כדי לקבל את קו ההחלטה במפורש ($x_2$ כפונקציה של $x_1$): י  



## י משוואה מדויקת של גבול ההחלטה: י  

י בוא נציג את המשוואה של הגבול הזה בצורה מדויקת: י  

$$
\hat{y}(x) = 0.5 \Rightarrow \beta_0 + \beta_1 x_1 + \beta_2 x_2 = 0.5
$$

י לאחר סידור מחדש נקבל: י  

$$
x_2 = -\frac{\beta_1}{\beta_2} x_1 - \frac{\beta_0 - 0.5}{\beta_2}
$$

י קו זה הוא בדיוק גבול ההחלטה של המודל. י  

י בוא נחשב ונציג את הקו במפורש: י  

⸻

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LinearRegression

# יצירת נתונים סינתטיים חדשים עם שני פיצ'רים
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0,
                           n_informative=2, n_clusters_per_class=1, random_state=42)

# אימון מודל רגרסיה לינארית
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# מקדמי המודל
beta_0 = lin_reg.intercept_
beta_1, beta_2 = lin_reg.coef_

# הדפסת המקדמים
(beta_0, beta_1, beta_2)

# יצירת גבול ההחלטה במפורש
x_values = np.linspace(X[:,0].min()-1, X[:,0].max()+1, 100)
y_values = -(beta_1/beta_2)*x_values - (beta_0 - 0.5)/beta_2

# ציור הנתונים עם גבול ההחלטה
plt.figure(figsize=(10, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Class 0 (Actual)', edgecolor='k')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Class 1 (Actual)', edgecolor='k')

# גבול ההחלטה
plt.plot(x_values, y_values, color='red', linewidth=2, label='Decision Boundary (y=0.5)')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Explicit Decision Boundary (Linear Regression)')
plt.legend()
plt.grid(True)
plt.show()
```



## י הסבר לגרף: י  

י הקו האדום הוא גבול ההחלטה של המודל: י  
- י כל נקודה על הקו מקיימת $f(x) = 0.5$, ולכן נמצאת בדיוק על הגבול בין הקבוצות. י  
- י כל נקודה מעל הקו האדום ($f(x) > 0.5$) תסווג לקבוצה $Class\ 1$. י  
- י כל נקודה מתחת לקו האדום ($f(x) < 0.5$) תסווג לקבוצה $Class\ 0$. י  

י הגבול לינארי כי השתמשנו במודל של רגרסיה לינארית. י  

י הפרמטרים של הקו הם: י  
- י $\beta_0$: האינטרספט (נקודת החיתוך עם הציר האנכי). י  
- י $\beta_1, \beta_2$: השיפועים עבור כל אחת מהתכונות. י
