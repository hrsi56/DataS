### Exercise 3

Draw a scatter with random arrays --> 90 values for x-points, y-points, colors and sizes

``` python

```

Add a ColorMap to the scatter

````python
import numpy as np

x = np.random.randint(90, size=(90))
y = np.random.randint(90, size=(90))
colors = np.random.randint(90, size=(90))
sizes = 10 * np.random.randint(90, size=(90))

plt.scatter(x,y,cmap='hsv',c=colors, s=sizes, alpha=0.6)
plt.show()
````

Draw a pie as in picture, with the same:<br>

<br>

Use the variables below:


````python
y = [10, 25, 35, 40, 60, 65]
l = ["Paper", "Plastic", "Glass", "Cartons", "Wood", "Steel"]
e = [0,0,0,0,0.2,0.4]
c = ["#f6ccff", "#ea80ff", "#d91aff", "#bf00e6", "#7f0099", "#550066"]

fig, ax = plt.subplots(figsize=(8, 8))
fig.patch.set_facecolor('black')  # Change the figure background to black

plt.pie(y, labels=[f"{la}\n{ya}" for la,ya in zip(l,y)] , explode=e,colors=c,autopct= '%.1f%%' ,  textprops= {'color': 'white', 'size' : 8} )
plt.legend(loc=(1.1,0.7) ,fontsize=8 ,title = 'Recycling').get_title().set_fontsize('8')
plt.savefig('fig.png',format='png', bbox_inches='tight', transparent= True)
````
