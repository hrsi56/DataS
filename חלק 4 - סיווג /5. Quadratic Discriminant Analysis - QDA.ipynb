# י שיעור $5$: ניתוח מבחין ריבועי ($Quadratic\ Discriminant\ Analysis$ – $QDA$) י  

י לאחר שלמדנו על שיטת $LDA$, נלמד כעת על שיטת סיווג כללית וגמישה יותר: $Quadratic\ Discriminant\ Analysis$ ($QDA$). י  

⸻

## י מהו $Quadratic\ Discriminant\ Analysis$ ($QDA$)? י  

י ההבדל המרכזי בין $LDA$ ל-$QDA$ הוא ההנחה לגבי מטריצת השונות ($\Sigma$): י  

- י ב-$LDA$: מטריצת קו-ואריאנס משותפת אחת לכל הקבוצות. י  
- י ב-$QDA$: לכל קבוצה יש מטריצת קו-ואריאנס משלה ($\Sigma_k$). י  

י במילים אחרות, $QDA$ יותר גמישה – מאפשרת לכל קבוצה להתאפיין בהתפלגות גאוסיאנית עם שונות וצורת פיזור שונה לחלוטין מהקבוצות האחרות. י  

⸻## י • ב-$Linear\ Discriminant\ Analysis$‏ ($LDA$) מניחים שכל הקבוצות חולקות אותה מטריצת קו-ואריאנס ($Covariance\ matrix$), ולכן מתקבל גבול החלטה לינארי. י  
## י • ב-$Quadratic\ Discriminant\ Analysis$ כל קבוצה $k$ מקבלת מטריצת קו-ואריאנס משלה, $\Sigma_k$. לפיכך, הגבול המפריד בין שתי קבוצות הופך לביטוי ריבועי, ומכאן השם "ריבועי". צורת גבול ההחלטה יכולה להיות עקומה מורכבת (כגון אליפסה), ולא ישר. י  

י הרעיון הבסיסי ב-$QDA$ הוא שימוש בחוק בייס, תחת ההנחה שכל קבוצה מתפלגת נורמלית רב-ממדית, אך עם קו-ואריאנס שונה. י  

⸻

# י המודל המתמטי י  

## י הנחת ההתפלגות י  

י נניח שיש לנו $K$ קבוצות ($classes$), ולכל קבוצה $k$ יש התפלגות נורמלית רב-ממדית ($multivariate\ normal$) במרחב בעל מימד $p$. כלומר, עבור וקטור תצפית $x \in \mathbb{R}^p$: י  

$$
f_k(x) = \frac{1}{(2\pi)^{p/2} |\Sigma_k|^{1/2}} \exp\left(-\frac{1}{2} (x-\mu_k)^T \Sigma_k^{-1} (x-\mu_k)\right)
$$

י כאשר: י  
1. י $\mu_k \in \mathbb{R}^p$ – וקטור הממוצע של הקבוצה $k$. י  
2. י $\Sigma_k \in \mathbb{R}^{p \times p}$ – מטריצת הקו-ואריאנס של הקבוצה $k$ (חיובית-מוגדרת, ולכן הפיכה). י  

י זו ההנחה המרכזית של $QDA$: לכל קבוצה יש התפלגות נורמלית משלה, כלומר פרמטרים שונים של ממוצע ($Mean$) וקו-ואריאנס ($Covariance$). י  

⸻

## י חוק בייס והחלטה י  

י במסגרת מודל הסתברותי, נרצה להעריך מהי ההסתברות ($Posterior\ probability$) שתצפית חדשה $x$ שייכת לקבוצה $k$: י  

$$
P(Y=k \mid X=x) = \frac{\pi_k \, f_k(x)}{\sum_{j=1}^K \pi_j \, f_j(x)}
$$

י כאשר $\pi_k = P(Y=k)$ היא ההסתברות המקדימה של הקבוצה $k$. י  
י בתהליך הסיווג, נבחר את הקבוצה $\hat{k}$ שנותנת את ההסתברות הגדולה ביותר: י  

$$
\hat{k}(x) = \underset{k}{\mathrm{argmax}}\; \pi_k \, f_k(x)
$$  

⸻

## י ביטוי גבול ההחלטה י  

י גבול ההחלטה בין שתי קבוצות $k$ ו-$l$: י  

$$
\log \frac{P(Y=k \mid X=x)}{P(Y=l \mid X=x)} = 0
$$  

י מתקבל: י  

$$
\log \frac{\pi_k \, f_k(x)}{\pi_l \, f_l(x)} = 0 \quad \Longleftrightarrow \quad \pi_k \, f_k(x) = \pi_l \, f_l(x)
$$  

י לאחר פתיחת הביטויים עבור $f_k(x)$ ו-$f_l(x)$, מתקבל: י  

$$
-\frac{1}{2}(x-\mu_k)^T \Sigma_k^{-1}(x-\mu_k)
+ \frac{1}{2}(x-\mu_l)^T \Sigma_l^{-1}(x-\mu_l)
+ \log \frac{\pi_k}{\pi_l} = 0
$$

י בשל הביטויים הריבועיים ב-$x$, גבול ההחלטה במרחב $\mathbb{R}^p$ הוא ריבועי. י  

⸻

## י מדוע "ריבועית" ($Quadratic$)? י  
- י ב-$LDA$, מניחים $\Sigma_1 = \Sigma_2 = \dots = \Sigma_K = \Sigma$. במקרה כזה, ההפרש בין הביטויים הופך ללינארי. י  
- י ב-$QDA$, לכל קבוצה מטריצה אחרת $\Sigma_k$, ולכן מתקבל ביטוי ריבועי ב-$x$. י  

⸻

# י שימושים ויתרונות-חסרונות י  

## י יתרונות: י  
- י גמישה יותר מ-$LDA$ – מתאימה למצבים שבהם אין סיבה להניח שוויון במטריצות הקו-ואריאנס בין הקבוצות. י  
- י יכולה להתמודד עם הבדלים משמעותיים בפיזור או בצורה בין הקבוצות. י  
- י מספקת הפרדה טובה יותר במקרים שבהם הגבול אינו לינארי. י  

## י חסרונות: י  
- י דורשת הערכה של יותר פרמטרים: מטריצת קו-ואריאנס $p \times p$ לכל קבוצה. י  
- י עלולה לסבול מ-$Overfitting$ כאשר מספר התצפיות קטן יחסית ל-$p$. י  
- י רגישה להערכה מדויקת של $\Sigma_k$, במיוחד כאשר מספר הדגימות נמוך. י  

⸻

# י סיכום עיקרי התהליך של $QDA$ י  

## י 1. אומדן פרמטרים: י  
- י מניחים לכל קבוצה $k$: התפלגות נורמלית עם $\mu_k$ ו-$\Sigma_k$. י  
- י מעריכים $\mu_k$ כממוצע התצפיות השייכות לקבוצה $k$. י  
- י מעריכים $\Sigma_k$ כמטריצת קו-ואריאנס של הקבוצה. י  
- י מחשבים $\pi_k$ לפי שיעור הקבוצה בכלל התצפיות. י  

## י 2. חישוב ההסתברות המאוחרת ($Posterior$): י  

$$
P(Y=k \mid X=x) = \frac{\pi_k \, f_k(x)}{\sum_{j=1}^K \pi_j \, f_j(x)}
$$

## י 3. סיווג: י  

$$
\hat{k}(x) = \underset{k}{\mathrm{argmax}}\; \pi_k \, f_k(x)
$$

## י 4. גבולות ההחלטה: י  
- י יחס ההסתברויות שווה $\Longrightarrow$ גבול ריבועי ב-$x$. י  

⸻

# י נקודות חשובות לסיכום י  
- י $QDA$ מניח התפלגות נורמלית שונה לכל קבוצה ($\mu_k, \Sigma_k$). י  
- י בניגוד ל-$LDA$, שבו $\Sigma$ זהה, ב-$QDA$ לכל קבוצה שונות משלה. י  
- י $QDA$ מתאים למצבים מורכבים, אך דורש כמות תצפיות גדולה לאמידה מדויקת. י  

י בקיצור, $Quadratic\ Discriminant\ Analysis$ הוא מודל הסתברותי-סטטיסטי גמיש, המשלב הסתברויות נורמליות עם גבולות החלטה ריבועיים. י  
⸻

## י השוואה בין $LDA$ ל-$QDA$: י  

| י תכונה י                                 | י $LDA$ י                  | י $QDA$ י                      |
|-------------------------------------------|----------------------------|-------------------------------|
| י מטריצת קו-ואריאנס י                    | י משותפת י                 | י שונה לכל קבוצה י           |
| י גבול החלטה י                           | י לינארי (ישר) י           | י ריבועי (אליפטי) י          |
| י מספר פרמטרים להערכה י                 | י קטן י                    | י גדול יותר י                |
| י ביצועים עם מעט נתונים י               | י טובה (מעט פרמטרים) י   | י פחות טובה, דורשת יותר נתונים י |
| י ביצועים כשאין נורמליות או שונות אחידה י | י פחות טוב י               | י עשוי להשתפר י               |

י הבחירה בין $LDA$ ל-$QDA$ תלויה בגודל הנתונים ובהנחות לגבי ההתפלגות של הקבוצות. י  

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Ellipse
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA

def plot_cov_ellipse(mean, cov, ax, n_std=2.0, edgecolor='black', lw=2, **kwargs):
    """ציור אליפסה המייצגת את אזור הקו-וריאנס סביב הממוצע."""
    vals, vecs = np.linalg.eigh(cov)
    order = vals.argsort()[::-1]
    vals, vecs = vals[order], vecs[:, order]

    theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))
    width, height = 2 * n_std * np.sqrt(vals)

    ellipse = Ellipse(
        xy=mean, width=width, height=height,
        angle=theta, edgecolor=edgecolor, facecolor='none', lw=lw, **kwargs
    )
    ax.add_patch(ellipse)

# הגדר seed לצורך שחזור
np.random.seed(0)

sns.set_style('whitegrid')
sns.set_context('talk')

# ------------------------------------------------------------------
# 1. יצירת נתונים לדוגמה, שתי קבוצות עם מטריצות קו-ואריאנס שונות
# ------------------------------------------------------------------

n_samples = 150

# ממוצעי שתי הקבוצות
mu_A = [2, 2]
mu_B = [-2, -2]

# נשים 2 מטריצות קו-ואריאנס שונות לשתי הקבוצות
# לדוגמה, קבוצה A עם קורלציה חיובית גבוהה
cov_A = [[1, 0.8],
         [0.8, 1]]
# וקבוצה B עם קורלציה שלילית
cov_B = [[1, -0.8],
         [-0.8, 1]]

# דגימה מקרית
classA = np.random.multivariate_normal(mu_A, cov_A, n_samples)
classB = np.random.multivariate_normal(mu_B, cov_B, n_samples)

X = np.vstack((classA, classB))
y = np.array([0]*n_samples + [1]*n_samples)


# -------------------------------------------
# 2. אימון מודל LDA ומודל QDA על אותם נתונים
# -------------------------------------------
lda_model = LDA()
qda_model = QDA()

lda_model.fit(X, y)
qda_model.fit(X, y)

# ---------------------------------------
# 3. בניית רשת של נקודות כדי לצייר גבולות
# ---------------------------------------
x_min, x_max = X[:, 0].min() - 2, X[:, 0].max() + 2
y_min, y_max = X[:, 1].min() - 2, X[:, 1].max() + 2

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)
grid_points = np.c_[xx.ravel(), yy.ravel()]

# תחזיות של שני המודלים על רשת הנקודות
Z_lda = lda_model.predict(grid_points).reshape(xx.shape)
Z_qda = qda_model.predict(grid_points).reshape(xx.shape)

# התחזית ההסתברותית (class=1) לשם ציור קווי ההפרדה (0.5)
Z_lda_proba = lda_model.predict_proba(grid_points)[:, 1].reshape(xx.shape)
Z_qda_proba = qda_model.predict_proba(grid_points)[:, 1].reshape(xx.shape)

# -----------------------------------
# 4. ציור - השוואת LDA מול QDA
# -----------------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# === LDA ===
ax = axes[0]
ax.set_title("LDA (יראיניל לובג )", fontsize=16)

# רקע הסיווג (Class 0 / Class 1)
ax.contourf(xx, yy, Z_lda, alpha=0.3, cmap=plt.cm.coolwarm)

# קו ההחלטה היכן ש-p(Class=1)=0.5
ax.contour(xx, yy, Z_lda_proba, levels=[0.5],
           colors='black', linestyles='--', linewidths=2)

# נתוני A
ax.scatter(classA[:,0], classA[:,1], c='blue', edgecolors='k', alpha=0.8, label='Class A')
# נתוני B
ax.scatter(classB[:,0], classB[:,1], c='red', edgecolors='k', alpha=0.8, label='Class B')

# ציור אליפסה של קו-ואריאנס סביב הממוצעים האמיתיים
plot_cov_ellipse(mu_A, cov_A, ax, edgecolor='blue', lw=2)
plot_cov_ellipse(mu_B, cov_B, ax, edgecolor='red', lw=2)

ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.legend(loc='upper left')

# === QDA ===
ax = axes[1]
ax.set_title("QDA (יעוביר לובג)", fontsize=16)

# רקע הסיווג (Class 0 / Class 1)
ax.contourf(xx, yy, Z_qda, alpha=0.3, cmap=plt.cm.coolwarm)

# קו ההחלטה היכן ש-p(Class=1)=0.5
ax.contour(xx, yy, Z_qda_proba, levels=[0.5],
           colors='black', linestyles='--', linewidths=2)

# נתוני A
ax.scatter(classA[:,0], classA[:,1], c='blue', edgecolors='k', alpha=0.8, label='Class A')
# נתוני B
ax.scatter(classB[:,0], classB[:,1], c='red', edgecolors='k', alpha=0.8, label='Class B')

# ציור אליפסה של הקו-ואריאנס (אותו הדבר, להמחשת "העולם האמיתי")
plot_cov_ellipse(mu_A, cov_A, ax, edgecolor='blue', lw=2)
plot_cov_ellipse(mu_B, cov_B, ax, edgecolor='red', lw=2)

ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
