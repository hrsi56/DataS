## subplot()

`subplots()` = multiple plots in one figure<br>
takes 3 arguments that describes the layout of the figure:

- First arg = Rows
- Second arg = Columns
- Third arg = Index

````python
x1 = [0,3,6,8]
y1 = [0,50,200,100]
print(x1)
print(y1)
````

````python
x2 = [0,2,4,6]
y2 = [10,25,15,30]
print(x2)
print(y2)
````

````python
#plot 1
#the figure has 1 row, 2 columns, and this plot is the first plot
plt.subplot(1, 2, 1)
plt.plot(x1,y1)

#plot 2
#the figure has 1 row, 2 columns, and this plot is the second plot
plt.subplot(1, 2, 2)
plt.plot(x2,y2)

plt.show()
````

````python
# Top of each other instead of side-by-side
#plot 1
plt.subplot(2, 1, 1)
plt.plot(x1,y1)

#plot 2
plt.subplot(2, 1, 2)
plt.plot(x2,y2)

plt.show()
````

````python
# Many Plots

#plot 1
plt.subplot(2, 3, 1)
plt.plot(x1,y1)

#plot 2
plt.subplot(2, 3, 2)
plt.plot(x2,y2)

#plot 3
plt.subplot(2, 3, 3)
plt.plot(x1,y1)


#plot 4
plt.subplot(2, 3, 4)
plt.plot(x1,y1)

#plot 5
plt.subplot(2, 3, 5)
plt.plot(x2,y2)

#plot 6
plt.subplot(2, 3, 6)
plt.plot(x1,y1)


plt.show()
````

## subplots()

<p style="font-size:20px"> The subplots() function is used to create a figure and a set of subplots.</p>

Syntax:<br>
``` python
matplotlib.pyplot.subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)
```

<u>Parameters: This method accept the following parameters that are described below:</u>
- nrows, ncols : These parameter are the number of rows/columns of the subplot grid.
- sharex, sharey : These parameter controls sharing of properties among x (sharex) or y (sharey) axes.
- squeeze : This parameter is an optional parameter and it contains boolean value with default as True.
- num: This parameter is the pyplot.figure keyword that sets the figure number or label.
- subplot_kwd: This parameter is the dict with keywords passed to the add_subplot call used to create each subplot.
- gridspec_kw: This parameter is the dict with keywords passed to the GridSpec constructor used to create the grid the subplots are placed on.


<u>Returns: This method return the following values</u>
- fig : This method return the figure layout.
- ax : This method return the axes.Axes object or array of Axes objects.

````python
x3 = [1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008]
y3 = [3267,2838,2871,2904,3663,2838,3399,2871,3102,2574,2541,2805,2838]

y4 = [500,354,264,231,660,561,300,297,305,363,396,297,198]
y5 = [v / y4[i] for i,v in enumerate(y3)]


# One "ax" (one Graph)
fig, ax = plt.subplots()

ax.plot(x3,y3, label = 'ProductRev')

plt.legend()

plt.show()
````

````python
# 3 "ax" (3 Graph)
fig, ax = plt.subplots(3)

ax[0].plot(x3,y3, label = 'ProductRev')
ax[1].plot(x3,y4, label = 'ProductSold')
ax[2].plot(x3,y5, label = 'AvgRevByPrd')

ax[0].legend()
ax[1].legend()
ax[2].legend()


plt.show()
````

````python
# 3 "ax" (3 Graph) beside each other
fig,ax = plt.subplots(nrows=1, ncols=3)


ax[0].plot(x3,y3, label = 'ProductRev')
ax[1].plot(x3,y4, label = 'ProductSold')
ax[2].plot(x3,y5, label = 'AvgRevByPrd')


ax[0].legend()
ax[1].legend()
ax[2].legend()


plt.show()
````

````python
# 6 "ax" (6 Graph) beside each other
fig, ax = plt.subplots(nrows=2, ncols=3)


ax[0,0].plot(x3,y3, label = 'ProductRev')
ax[0,1].plot(x3,y4, label = 'ProductSold')
ax[0,2].plot(x3,y5, label = 'AvgRevByPrd')
ax[1,0].plot(x3,y3, label = 'ProductRev')
ax[1,1].plot(x3,y4, label = 'ProductSold')
ax[1,2].plot(x3,y5, label = 'AvgRevByPrd')


plt.show()
````

### Increasing fig size - one way

````python
# 6 "ax" (6 Graph) beside each other
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 15))

ax[0,0].plot(x3,y3, label = 'ProductRev')
ax[0,1].plot(x3,y4, label = 'ProductSold')
ax[0,2].plot(x3,y5, label = 'AvgRevByPrd')
ax[1,0].plot(x3,y3, label = 'ProductRev')
ax[1,1].plot(x3,y4, label = 'ProductSold')
ax[1,2].plot(x3,y5, label = 'AvgRevByPrd')


plt.show()
````

### Increasing fig size - Another way

````python
# 6 "ax" (6 Graph) beside each other
fig, ax = plt.subplots(nrows=2, ncols=3)

ax[0,0].plot(x3,y3, label = 'ProductRev')
ax[0,1].plot(x3,y4, label = 'ProductSold')
ax[0,2].plot(x3,y5, label = 'AvgRevByPrd')
ax[1,0].plot(x3,y3, label = 'ProductRev')
ax[1,1].plot(x3,y4, label = 'ProductSold')
ax[1,2].plot(x3,y5, label = 'AvgRevByPrd')

fig.set_figheight(15)
fig.set_figwidth(15)


plt.show()
````

## hist()

````python
# Let's create an array using NumPy with Normal Data Distribution.
# For students grades, 500 values, concentrate around 80, and the standard deviation is 10.
import numpy as np
grades = np.random.normal(80, 10, 500)
print(grades)
````

### Sorting pie

````python
# Start Angle
# Notice that the default start angle is at the x-axis also known as zero (0) angle,
# in our case it starts with "Pokimon" (red), next "Power Rangers" (green), "Ninja Turtles" (yello), "Transformers" (blue), it's the order of the array starts from index -1!!!

y = [25, 10, 35, 40]
l = ["Transformers", "Ninja Turtles", "Power Rangers", "Pokimon"]

plt.pie(y, labels = l, autopct='%1.1f%%')
plt.show()
````

Let's say we wanted to change the order by size (desc), all we need to do is to sort the order of the array

````python
# BREAKDOWN !


# Initial data
y = [25, 10, 35, 40]
l = ["Transformers", "Ninja Turtles", "Power Rangers", "Pokimon"]

# zip() --> create list of tupples
list(zip(y, l))
````

````python
# Sorting 'y' in ascending order and arranging 'l' to correspond to 'y'
sorted(zip(y, l))
````

````python
# Creating 2 sorted lists
ys = [v[0] for v in sorted(zip(y, l))]
ls = [v[1] for v in sorted(zip(y, l))]

print(ys)
print(ls)
````

````python
# Creating 2 sorted lists, but this time the lable with data value
ys = [v[0] for v in sorted(zip(y, l))]
ls = [f'''{v[1]} \n {str(v[0])} hours''' for v in sorted(zip(y, l))]

print(ys)
print(ls)
````

````python
plt.pie(ys, labels = ls, autopct='%1.1f%%')
plt.show()

# Remember! the default start angle is at the x-axis also known as zero (0) angle,
# in our case, now,  it starts with "Pokimon" (red), next "Power Rangers" (green)...
````

````python
# Start Angle
# startangle parameter is defined with an angle in degrees, default angle is 0

# Let's start from 90 degrees (in our case "Pokimon" (red))

plt.pie(ys, labels = ls, autopct='%1.1f%%', startangle = 90)
plt.show()
````

````python
# Start Angle
# start from 270 degrees (in our case "Pokimon" (red))

plt.pie(ys, labels = ls, autopct='%1.1f%%', startangle = 270)
plt.show()
````

````python
# Start Angle
# start from 180 degrees (in our case "Pokimon" (red))

plt.pie(ys, labels = ls, autopct='%1.1f%%', startangle = 180)
plt.show()
````

````python
# Start Angle
# start from 0 degrees (default) (in our case "Pokimon" (red))

plt.pie(ys, labels = ls, autopct='%1.1f%%', startangle = 0)
plt.show()
````

````python
# Explode
# The explode parameter must be an array with one value for each elemnt
# Each value represents how far from the center each element is displayed

# "Pokimon" is pulled 0.3 from the center
e = [0,0,0.1,0.3]

plt.pie(ys, labels = ls, autopct='%1.1f%%', explode = e)
plt.show()
````

````python
# Color

c = ["blue", "hotpink", "y", "#4CAF50"]

plt.pie(ys, labels = ls, autopct='%1.1f%%', explode = e, colors = c)
plt.show()
````

````python
# Legend

plt.pie(ys, labels = ls, autopct='%1.1f%%', explode = e, colors = c, textprops=dict(color="white"))

plt.legend()

plt.show()

# Legend overlaps with the pie chart !!!
````

````python
# Legend position

plt.pie(ys, labels = ls, autopct='%1.1f%%', explode = e, colors = c, textprops=dict(color="white"))


plt.legend(title = "Tv Series", loc="upper left")

plt.show()


# The legend STILL overlaps with the pie chart !!!
````

````python
# Legend position

# e, bbox_to_anchor=(1, 1) moves the legend outside the pie chart to the upper left corner.
# Adjust the bbox_to_anchor values to find a position that works best for your chart.

plt.pie(ys, labels = ls, autopct='%1.1f%%', explode = e, colors = c, textprops=dict(color="white"))


plt.legend(title = "Tv Series", bbox_to_anchor=(1, 1))

plt.show()
````
